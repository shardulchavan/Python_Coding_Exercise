```python
#11 Check if a given string is symmetrical and palindrome or not.
str1=input("Enter string to check whether it is palindorme & symmetrical:")
len_half_str1=int(len(str1)/2)
if len(str1)%2==0:
  first_half=str1[:len_half_str1]
  second_half=str1[len_half_str1:]
else:
  first_half=str1[0:len_half_str1+2]
  second_half=str1[len_half_str1-1:]
print(first_half,'',second_half)
if first_half==second_half:
  print('symmetrcial')
else:
  print('asymmetrical')
#print(second_half[::-1])
if first_half==second_half[::-1]:
  print('Palindrome')
else:
  print('Not Palindrome')
```

```python
#12 Write a Python program to find the n-th term in a Fibonacci series using recursion. 
def fibo(num):
  if num<0:
    return ('Invalid option')
  elif num==1:
    return 0
  elif num==2:
    return 1
  else:
    return fibo(num-1)+fibo(num-2)
num=int(input('Find fibonacci series of following no.: '))
result=fibo(num)
print(result)
```

```python
# # 13Write a Python program to find the longest substring between two given strings. You may not 
# use any package except the substring function but you can make use of recursive function. Here 
# is the way you will be graded:
import time
start_time = time.time()
def main():
  print("--- %s seconds ---" % (time.time() - start_time))
def longest_commonsubstrng(str1,str2):
  len1=len(str1)
  len2=len(str2)
  counter=[[0]*(len2+1)for  x in range(len1+1)]
  longest=0
  lcs_set=set()
  for i in  range(len1):
    for j in range(len2):
      if str1[i]==str2[j]:
        c=counter[i][j]+1
        counter[i+1][j+1]=c
        if c>longest:
          lcs_set=set()
          longest=c
          lcs_set.add(str1[i-c+1:i+1])
        elif c==longest:
          lcs_set.add(str1[i-c+1:i+1])
  return lcs_set
string1=input('Enter first string')
string2=input('Enter second string')
result=longest_commonsubstrng(string1,string2)
for res in result:
  print('Longest Common Substring:',res)
main()
```
